```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# First plot
np.random.seed(613)
x = np.arange(50)
y = np.random.randint(0,100,50)

fig, ax = plt.subplots(figsize = (5,3))
# ax.scatter(x,y)
# ax.bar(x,y)
# ax.plot(x,y)
ax.hist(y)
```

```python
# Second plot

font1 = {'family':'sans-serif', 'color':'blue', 'size':20}
font2 = {'family':'monospace','color':'green', 'size':14}

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y)

ax.set_title('Total growth over time', fontdict=font1)
ax.set_ylabel('Total growth', fontdict = font2)
ax.set_xlabel('Years since start', fontdict = font2)
plt.show()
```

```python
# Scatter plot

fig, ax = plt.subplots(figsize = (5,3))
ax.scatter(x,y,
           marker = '*',
           color = 'indigo')
```

```python
# Plot with grid

fig, ax = plt.subplots(figsize = (5,3))
ax.plot(x,y,
           marker = '*',
           color = 'indigo',
           linestyle = '--',
           linewidth = 2,
           markersize = 12,
           markeredgecolor = '#fa9359',
           markerfacecolor = '#000000')

ax.grid(color='blue',
        linewidth = 2,
        linestyle = 'dotted')
```
